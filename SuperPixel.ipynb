```python
from __future__ import absolute_import, division, print_function, unicode_literals
import numpy as np
import os
import matplotlib.pyplot as plt
import time
from skimage.segmentation import slic, quickshift, watershed, felzenszwalb
from skimage.segmentation import mark_boundaries
from skimage.util import img_as_float
from skimage import io
from skimage.color import rgb2gray
from skimage.filters import sobel
import matplotlib.pyplot as plt
import cv2
import warnings; warnings.simplefilter('ignore')
```

```python
image_path = 'E:\\PhD\\jpg2\\000041_20050607145750A.jpg'
image = img_as_float(io.imread(image_path))
```

```python
plt.imshow(image)
```

Linear Filter

```python
for numSegments in (100, 200, 300, 400, 500):
        segments = slic(image, n_segments = 400, sigma = 5, compactness=10.0)

        fig = plt.figure("Superpixels -- %d segments" % (numSegments))
        ax = fig.add_subplot(1, 1, 1)
        ax.imshow(mark_boundaries(image, segments))
        plt.axis("off")
 
        plt.show()
```

```python
for numCompact in (10, 100):
        segments = slic(image, n_segments = numSegments, sigma = 0.5, compactness=numCompact)

        fig = plt.figure("Superpixels -- %d segments" % (numSegments))
        ax = fig.add_subplot(1, 1, 1)
        ax.imshow(mark_boundaries(image, segments))
        plt.axis("off")
 
        plt.show()
```

```python
for numSegments in (100, 200, 300):
        segments = felzenszwalb(image, scale=numSegments, sigma=0.5, min_size=50)
        fig = plt.figure("Superpixels -- %d segments" % (numSegments))
        ax = fig.add_subplot(1, 1, 1)
        ax.imshow(mark_boundaries(image, segments))
        plt.axis("off")
 
        plt.show()
```
